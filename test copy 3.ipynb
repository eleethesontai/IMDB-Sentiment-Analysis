```python
import pandas as pd
import numpy as np
from tokenizers import ByteLevelBPETokenizer
from itertools import combinations
```

```python
samples = pd.read_csv('Reviews.csv')
```

```python
tokenizer = ByteLevelBPETokenizer()
tokenizer.train_from_iterator(samples['Review'], vocab_size=4096, min_frequency=256, special_tokens=["[PAD]", "[MASK]"])
```

```python
vocabulary = tokenizer.get_vocab()
dict(sorted(vocabulary.items(), key=lambda item: item[1]))
```
